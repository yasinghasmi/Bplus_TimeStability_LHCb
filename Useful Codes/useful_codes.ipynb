```python
# Check percentage of duplicates

from collections import Counter

#for filename in fill_files:
# Load the tree
tree = r.TChain("ST-b2oc")
tree.Add("filename")

# Use an appropriate identifier – typically "event" or similar
event_ids = []

# Replace 'event' with the actual branch name used to identify events
for entry in tree:
    event_ids.append(getattr(entry, "event"))  # Replace 'event' with actual name if different

# Count duplicates
counts = Counter(event_ids)
num_duplicates = sum(1 for count in counts.values() if count > 1)

fraction_duplicates = num_duplicates/tree.GetEntries()
print(f"Fraction of duplicated events: {fraction_duplicates}")
```
