# Plot of Branching Ratio Value

This section provides code to plot the values obtained for the Branching Fraction Ratio per Blocks, with respect to the current PDG world average and other experiments such as BaBar and Belle (calculation for these two in the first cell of this section). 


```python
import math

print("=== LITERATURE BRANCHING RATIO CALCULATION ===")
print()

# BELLE measurements from image
print("BELLE (2021) Measurements:")
# J/ψK+: 1.032 ± 0.007 ± 0.024 (×10^-3)
BF_JpsiK_Belle = 1.032e-3
BF_JpsiK_Belle_stat = 0.007e-3
BF_JpsiK_Belle_syst = 0.024e-3
BF_JpsiK_Belle_total = math.sqrt(BF_JpsiK_Belle_stat**2 + BF_JpsiK_Belle_syst**2)

# D0π+: 4.53 ± 0.02 ± 0.15 (×10^-3)
BF_Dpi_Belle = 4.53e-3
BF_Dpi_Belle_stat = 0.02e-3
BF_Dpi_Belle_syst = 0.15e-3
BF_Dpi_Belle_total = math.sqrt(BF_Dpi_Belle_stat**2 + BF_Dpi_Belle_syst**2)

print(f"  BF(B+ → J/ψK+) = ({BF_JpsiK_Belle*1e3:.3f} ± {BF_JpsiK_Belle_stat*1e3:.3f} ± {BF_JpsiK_Belle_syst*1e3:.3f}) × 10^-3")
print(f"  BF(B+ → D0π+) = ({BF_Dpi_Belle*1e3:.2f} ± {BF_Dpi_Belle_stat*1e3:.2f} ± {BF_Dpi_Belle_syst*1e3:.2f}) × 10^-3")
print(f"  Total errors: ±{BF_JpsiK_Belle_total*1e3:.3f} and ±{BF_Dpi_Belle_total*1e3:.3f}")
print()

# BABAR measurements from image
print("BABAR (2005) Measurements:")
# J/ψK+: 1.061 ± 0.015 ± 0.048 (×10^-3)
BF_JpsiK_BaBar = 1.061e-3
BF_JpsiK_BaBar_stat = 0.015e-3
BF_JpsiK_BaBar_syst = 0.048e-3
BF_JpsiK_BaBar_total = math.sqrt(BF_JpsiK_BaBar_stat**2 + BF_JpsiK_BaBar_syst**2)

# D0π+: 4.90 ± 0.07 ± 0.22 (×10^-3)
BF_Dpi_BaBar = 4.90e-3
BF_Dpi_BaBar_stat = 0.07e-3
BF_Dpi_BaBar_syst = 0.22e-3
BF_Dpi_BaBar_total = math.sqrt(BF_Dpi_BaBar_stat**2 + BF_Dpi_BaBar_syst**2)

print(f"  BF(B+ → J/ψK+) = ({BF_JpsiK_BaBar*1e3:.3f} ± {BF_JpsiK_BaBar_stat*1e3:.3f} ± {BF_JpsiK_BaBar_syst*1e3:.3f}) × 10^-3")
print(f"  BF(B+ → D0π+) = ({BF_Dpi_BaBar*1e3:.2f} ± {BF_Dpi_BaBar_stat*1e3:.2f} ± {BF_Dpi_BaBar_syst*1e3:.2f}) × 10^-3")
print(f"  Total errors: ±{BF_JpsiK_BaBar_total*1e3:.3f} and ±{BF_Dpi_BaBar_total*1e3:.3f}")
print()


def calculate_ratio_with_errors(BF_numerator, BF_numerator_err, BF_denominator, BF_denominator_err, experiment):
    """
    Calculate BF(D0π+)/BF(J/ψK+) with error propagation
    For ratio R = A/B: δR/R = √((δA/A)² + (δB/B)²)
    """
    ratio = BF_numerator / BF_denominator
    
    rel_err_num = BF_numerator_err / BF_numerator
    rel_err_den = BF_denominator_err / BF_denominator
    rel_err_ratio = math.sqrt(rel_err_num**2 + rel_err_den**2)
    
    ratio_err = ratio * rel_err_ratio
    
    return ratio, ratio_err, rel_err_ratio

print("=== BRANCHING RATIO RESULTS ===")

# BELLE calculation
Belle_ratio, Belle_ratio_err, Belle_rel_err = calculate_ratio_with_errors(
    BF_Dpi_Belle, BF_Dpi_Belle_total, 
    BF_JpsiK_Belle, BF_JpsiK_Belle_total, 
    "Belle"
)

print(f"BELLE:")
print(f"  BF(D0π+)/BF(J/ψK+) = {BF_Dpi_Belle*1e3:.2f}/{BF_JpsiK_Belle*1e3:.3f} = {Belle_ratio:.3f}")
print(f"  Error propagation: √(({BF_Dpi_Belle_total/BF_Dpi_Belle:.4f})² + ({BF_JpsiK_Belle_total/BF_JpsiK_Belle:.4f})²) = {Belle_rel_err:.4f}")
print(f"  Final result: {Belle_ratio:.3f} ± {Belle_ratio_err:.3f}")
print()

# BABAR calculation
BaBar_ratio, BaBar_ratio_err, BaBar_rel_err = calculate_ratio_with_errors(
    BF_Dpi_BaBar, BF_Dpi_BaBar_total,
    BF_JpsiK_BaBar, BF_JpsiK_BaBar_total,
    "BaBar"
)

print(f"BABAR:")
print(f"  BF(D0π+)/BF(J/ψK+) = {BF_Dpi_BaBar*1e3:.2f}/{BF_JpsiK_BaBar*1e3:.3f} = {BaBar_ratio:.3f}")
print(f"  Error propagation: √(({BF_Dpi_BaBar_total/BF_Dpi_BaBar:.4f})² + ({BF_JpsiK_BaBar_total/BF_JpsiK_BaBar:.4f})²) = {BaBar_rel_err:.4f}")
print(f"  Final result: {BaBar_ratio:.3f} ± {BaBar_ratio_err:.3f}")
print()

print("=== FINAL VARIABLES ===")
print(f"Belle_BF_ratio = {Belle_ratio:.3f}")
print(f"Belle_BF_ratio_err = {Belle_ratio_err:.3f}")
print(f"BaBar_BF_ratio = {BaBar_ratio:.3f}")
print(f"BaBar_BF_ratio_err = {BaBar_ratio_err:.3f}")
print()

# Store as variables for further use
Belle_BF_ratio = Belle_ratio
Belle_BF_ratio_err = Belle_ratio_err
BaBar_BF_ratio = BaBar_ratio
BaBar_BF_ratio_err = BaBar_ratio_err
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Enter your actual branching fraction values for each block
block5_BF_ratio = 4.365835524758542
block5_BF_error = 0.06661310652

block6_BF_ratio = 4.233966881889545
block6_BF_error = 0.06620322986

block7_BF_ratio = 4.51290114225437    
block7_BF_error = 0.0691058306

block8_BF_ratio = 4.465345219414248
block8_BF_error = 0.06934405474

print("=== BLOCK VALUES ===")
print(f"Block 5: {block5_BF_ratio:.2f} ± {block5_BF_error:.2f}")
print(f"Block 6: {block6_BF_ratio:.2f} ± {block6_BF_error:.2f}")
print(f"Block 7: {block7_BF_ratio:.2f} ± {block7_BF_error:.2f}")
print(f"Block 8: {block8_BF_ratio:.2f} ± {block8_BF_error:.2f}")
print()

# World average from the image
world_avg_value = 4.48
world_avg_error = 0.12

print("Creating comparison plot with all blocks...")

# Create figure
fig, ax = plt.subplots(figsize=(12, 10))

# Data for plotting (y-positions for vertical layout with grouped LHCb blocks)
experiments = ['BaBar\n(2005)', 'Block 5\n(LHCb)', 'Block 6\n(LHCb)', 'Block 7\n(LHCb)', 'Block 8\n(LHCb)', 'Belle\n(2021)']
values = [BaBar_BF_ratio, block5_BF_ratio, block6_BF_ratio, block7_BF_ratio, block8_BF_ratio, Belle_BF_ratio]
errors = [BaBar_BF_ratio_err, block5_BF_error, block6_BF_error, block7_BF_error, block8_BF_error, Belle_BF_ratio_err]
y_positions = [2, 3, 3.5, 4, 4.5, 5.5]  # BaBar isolated at 1, LHCb blocks grouped 3-4.5, Belle isolated at 6.5

# Colors for each measurement
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']  # Different color for each
labels = ['BaBar', 'Block 5', 'Block 6', 'Block 7', 'Block 8', 'Belle']

# Plot world average as shaded region (horizontal band)
world_lower = world_avg_value - world_avg_error
world_upper = world_avg_value + world_avg_error

# Shaded region for world average uncertainty
ax.axvspan(world_lower, world_upper, alpha=0.3, color='darkgray', 
           label=f'World Average: {world_avg_value:.2f} ± {world_avg_error:.2f}', zorder=1)

# Central line for world average value
ax.axvline(world_avg_value, color='darkgray', linewidth=2, linestyle='-', zorder=2)

# Plot individual measurements with error bars
for i, (exp, val, err, y_pos, color, label) in enumerate(zip(experiments, values, errors, y_positions, colors, labels)):
    # Make LHCb blocks use square markers to distinguish from Belle/BaBar circles
    if 'Block' in label:
        marker_style = 's'  
        marker_size = 7
        alpha = 0.9  #
    else:
        marker_style = 'o'  
        marker_size = 8
        alpha = 1.0
    
    # Error bars (horizontal)
    ax.errorbar(val, y_pos, xerr=err, fmt=marker_style, markersize=marker_size, 
                color=color, capsize=5, capthick=2, linewidth=2, alpha=alpha,
                label=f'{label}: {val:.2f} ± {err:.2f}', zorder=3)

# Add visual separators to emphasize grouping
ax.axhline(y=2.2, color='lightgray', linestyle='--', alpha=0.5, linewidth=1)  
ax.axhline(y=5.4, color='lightgray', linestyle='--', alpha=0.5, linewidth=1)  

# Add group labels
ax.text(4.01, 3.75, 'LHCb Analysis\n(This Work)', fontsize=10, fontweight='bold', 
        verticalalignment='center', bbox=dict(boxstyle='round,pad=0.3', facecolor='lightblue', alpha=0.7))

# Customize the plot
ax.set_ylim(0.5, 7.0)  # Adjusted for new y-positions
ax.set_xlim(4.0, 5.0)  # Adjust range as needed for your values
ax.set_xlabel('BF(B⁺ → D⁰π⁺) / BF(B⁺ → J/ψK⁺)', fontsize=14, fontweight='bold')
ax.set_ylabel('Experiment / Block', fontsize=14, fontweight='bold')
ax.set_title('Branching Fraction Ratio Comparison by Block\nB⁺ → D⁰π⁺ / B⁺ → J/ψK⁺', 
             fontsize=16, fontweight='bold', pad=20)

# Set y-tick labels
ax.set_yticks(y_positions)
ax.set_yticklabels(experiments, fontsize=12)

# Add grid
ax.grid(True, alpha=0.3, axis='x')

# Add legend
ax.legend(loc='upper right', fontsize=10, framealpha=0.9)

# Add text box with summary
textstr = f'''Summary:
Belle:     {Belle_BF_ratio:.2f} ± {Belle_BF_ratio_err:.2f}
Block 8:   {block8_BF_ratio:.2f} ± {block8_BF_error:.2f}
Block 7:   {block7_BF_ratio:.2f} ± {block7_BF_error:.2f}
Block 6:   {block6_BF_ratio:.2f} ± {block6_BF_error:.2f}
Block 5:   {block5_BF_ratio:.2f} ± {block5_BF_error:.2f}
BaBar:     {BaBar_BF_ratio:.2f} ± {BaBar_BF_ratio_err:.2f}
PDG:     {world_avg_value:.2f} ± {world_avg_error:.2f}'''

# Add text box
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
ax.text(0.02, 0.98, textstr, transform=ax.transAxes, fontsize=9,
        verticalalignment='top', bbox=props)

# Adjust layout and show
plt.tight_layout()
plt.show()

print()
print("=== BLOCK-BY-BLOCK COMPARISON SUMMARY ===")
print(f"Belle (2021):      {Belle_BF_ratio:.2f} ± {Belle_BF_ratio_err:.2f}")
print(f"Block 8 (LHCb):    {block8_BF_ratio:.2f} ± {block8_BF_error:.2f}")
print(f"Block 7 (LHCb):    {block7_BF_ratio:.2f} ± {block7_BF_error:.2f}")
print(f"Block 6 (LHCb):    {block6_BF_ratio:.2f} ± {block6_BF_error:.2f}")
print(f"Block 5 (LHCb):    {block5_BF_ratio:.2f} ± {block5_BF_error:.2f}")
print(f"BaBar (2005):      {BaBar_BF_ratio:.2f} ± {BaBar_BF_ratio_err:.2f}")
print(f"World Average:     {world_avg_value:.2f} ± {world_avg_error:.2f}")
print()

# Calculate differences from world average for all blocks
print("Differences from World Average:")
print(f"Belle:  {abs(Belle_BF_ratio - world_avg_value):.2f} ({abs(Belle_BF_ratio - world_avg_value)/world_avg_error:.1f}σ)")
print(f"Block 8: {abs(block8_BF_ratio - world_avg_value):.2f} ({abs(block8_BF_ratio - world_avg_value)/world_avg_error:.1f}σ)")
print(f"Block 7: {abs(block7_BF_ratio - world_avg_value):.2f} ({abs(block7_BF_ratio - world_avg_value)/world_avg_error:.1f}σ)")
print(f"Block 6: {abs(block6_BF_ratio - world_avg_value):.2f} ({abs(block6_BF_ratio - world_avg_value)/world_avg_error:.1f}σ)")
print(f"Block 5: {abs(block5_BF_ratio - world_avg_value):.2f} ({abs(block5_BF_ratio - world_avg_value)/world_avg_error:.1f}σ)")
print(f"BaBar:  {abs(BaBar_BF_ratio - world_avg_value):.2f} ({abs(BaBar_BF_ratio - world_avg_value)/world_avg_error:.1f}σ)")
```
